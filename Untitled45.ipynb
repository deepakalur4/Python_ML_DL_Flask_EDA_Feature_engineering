```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(r"https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
```

```python
df.head()
```

```python
df.rename(columns={"medv":"Price"},inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=pd.DataFrame(df.iloc[:,-1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test)
```

```python
from tensorflow import keras
```

```python
classifer=keras.models.Sequential()
```

```python
classifer.add(keras.layers.Dense(units=32,kernel_initializer="normal",activation="relu",input_dim=13))
```

```python
classifer.add(keras.layers.Dense(units=32,kernel_initializer="normal",activation="relu"))
```

```python
classifer.add(keras.layers.Dense(units=1, kernel_initializer= "normal",activation= "relu"))
```

```python
classifer.compile(optimizer="adam",loss="mean_absolute_error",metrics=["mean_absolute_error"])
```

```python
classifer.summary()
```

```python
model=classifer.fit(X_train,y_train,validation_split=0.33,batch_size=2,epochs=10)
```

```python
y_pred=classifer.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_pred,y_test))
```

```python
def build_model(hp):
  model = keras.Sequential()
  for i in range(hp.Int("num_layers",2,20)):
    model.add(keras.layers.Dense(
        units=hp.Int('units', min_value=20,max_value=500,step=10),
        activation='relu'))
  model.add(keras.layers.Dense(1, activation='relu'))
  model.compile(loss='mse')
  return model
```

```python
tuner = keras_tuner.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=5)
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
best_model = tuner.get_best_models()[0]
```
