```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
```

```python
df.head()
```

```python
df.rename(columns={"medv":"Price"},inplace=True)
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),cmap="viridis")
```

```python
height=df[["rm"]]
weight=df[["Price"]]
```

```python
type(height)
```

```python
df=pd.DataFrame()
```

```python
df["height"]=height
df["weight"]=weight
```

```python
sns.pairplot(df)
```

```python
plt.scatter(x=df["height"],y=df["weight"])
```

```python
sns.regplot(x=df["height"],y=df["weight"])
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x=df[["height"]] #independet feature
y=df["weight"] #dependent feature
```

```python
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression(n_jobs=-1)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# plt.scatter(x=X_train,y=y_train)
sns.scatterplot(x=y_test,y=reg.predict(X_test),hue=y_test)
```

```python
y_pred=reg.predict(X_test)
```

```python
res=y_test-y_pred
```

```python
sns.distplot(res,kde=True)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
```

```python
print(mse,mae,rmse)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
```

```python
print(score)
```

```python
import statsmodels.api as sn
```

```python
model=sn.OLS(y_train,X_train).fit()
```

```python
model.predict(X_test)
```

```python
print(model.summary())
```
