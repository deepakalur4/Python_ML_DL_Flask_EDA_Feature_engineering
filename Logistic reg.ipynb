```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplot inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
df.head()
```

```python
df["species"].unique()
```

```python
df.isnull().sum()
```

```python
df=df[df["species"]!="setosa"]
```

```python
df["species"].unique()
```

```python
df["species"]=df["species"].map({"versicolor":0,"virginica":1})
```

```python
df.head()
```

```python
df["species"].unique()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
los=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={"penalty":["l1","l2","elasticnet"],"C":[1,2,3,4,5,6,10,20,30,40,50,60],"max_iter":[100,200,300]}
```

```python
clasifer_reg=GridSearchCV(los,param_grid=parameters,scoring="accuracy",cv=5)
```

```python
clasifer_reg.fit(X_train,y_train)
```

```python
clasifer_reg.best_params_
```

```python
clasifer_reg.best_score_
```

```python
y_pred=clasifer_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
