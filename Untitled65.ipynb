```python
import spacy.cli
spacy.cli.download("en_core_web_lg")
nlp = spacy.load("en_core_web_lg")
```

```python
doc=nlp("Dr. stage love sto eat the vadapav of mumbai. and he is currently love's the chaat of delhi")
```

```python
for sent in doc.sents:
  print(sent)
  for word in sent:
    print(word)
```
