```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/spam.csv")
```

```python
df.head()
```

```python
df.groupby(["Category"]).describe()
```

```python
df["Category"].unique()
```

```python
df["Category"].apply(lambda x:1 if x=="spam" else 0)
```

```python
df["spam"]=df["Category"].apply(lambda x:1 if x=="spam" else 0)
```

```python
X=df["Message"]
y=df["spam"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
x_count=vectorizer.fit_transform(X_train)
```

```python
x_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(x_count,y_train)
```

```python
x_test=vectorizer.transform(X_test)
model.score(x_test,y_test)
```
