```python
import pandas as pd,numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/nlp-tutorials/main/11_bag_of_n_grams/Fake_Real_Data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df["label"].value_counts()
```

```python
df["label"]=df["label"].apply(lambda x: 1 if x=="Real" else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df["Text"], df["label"], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf=Pipeline([
    ("vec",CountVectorizer(ngram_range=(1,3))),
    ("model",KNeighborsClassifier(n_neighbors=10,metric="euclidean"))

])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
clf=Pipeline([
    ("vec",CountVectorizer(ngram_range=(1,3))),
    ("model",KNeighborsClassifier(n_neighbors=10,metric="cosine"))

])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=Pipeline([
    ("vec",CountVectorizer(ngram_range=(3,3))),
    ("model",RandomForestClassifier())

])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
