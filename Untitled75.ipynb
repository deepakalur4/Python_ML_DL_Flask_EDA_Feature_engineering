```python
import fasttext
```

```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/nlp-tutorials/main/17_fasttext/Cleaned_Indian_Food_Dataset.csv")
```

```python
df["TranslatedInstructions"][0]
```

```python
import re
```

```python
re.sub(r"[^\w\s]",' ',df["TranslatedInstructions"][0])
```

```python
def preprocess(text):
  step_1=re.sub(r"[^\w\s]",' ',text)
  step_2=re.sub(r"[  \n]+",' ',text)
  return step_2.strip().lower()
```

```python
preprocess(df["TranslatedInstructions"][0])
```

```python
df["TranslatedInstructions"]=df["TranslatedInstructions"].map(preprocess)
```

```python
df["TranslatedInstructions"][2]
```

```python
df.to_csv("deep.txt",columns=["TranslatedInstructions"],header=None,index=False)
```

```python
model=fasttext.train_unsupervised("deep.txt")
```

```python
model.get_nearest_neighbors("halwa")
```
