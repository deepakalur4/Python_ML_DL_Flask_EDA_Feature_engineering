```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/crncck/Fake-News-Detection/main/train.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
import tensorflow
```

```python
tensorflow.__version__
```

```python
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import models
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=5000
```

```python
one_hot("Deepoak",voc_size)
```

```python
messages=x.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
messages.head()
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages["title"][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word not in stopwords.words("english")]
  review=' '.join(review)
  corpus.append(review)
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in corpus]
```

```python
one_hot_repr[2]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
docs=pad_sequences(one_hot_repr,maxlen=20, padding='pre')
```

```python
docs[0]
```

```python
model=models.Sequential([
    layers.Embedding(5000,40,input_length=20),
    layers.LSTM(100),
    layers.Dense(units=1,activation="sigmoid"),
])
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.summary()
```

```python
x=docs
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5,batch_size=64)
```

```python
y_pred=model.predict(X_test)
```
