```python
from tensorflow.keras import datasets
```

```python
dataset=datasets.fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(train_iamges,train_labels),(test_images,test_labels)=dataset
```

```python
plt.imshow(train_iamges[0])
```

```python
train_iamges=train_iamges/255.0
test_images=test_images/255.0
```

```python
test_images.shape
```

```python
train_iamges=train_iamges.reshape(60000, 28, 28,1)
test_images=test_images.reshape(10000, 28, 28,1)
```

```python
model=keras.models.Sequential([
    keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation="relu",input_shape=(28,28,1)),
    keras.layers.Flatten(),
    keras.layers.Dense(10,activation="relu",),
    keras.layers.Dense(10,activation="softmax")
])

model.compile(optimizer="Adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(train_iamges,train_labels,batch_size=32,epochs=2)
```

```python
def build_model(hp):
  model=keras.models.Sequential([
      keras.layers.Conv2D(filters=hp.Int("conv_1_filter",min_value=32,max_value=128,step=16),
                          kernel_size=hp.Choice("conv_1_kernel",values=[3,5,8]),
                          activation="relu",
                          input_shape=(28,28,1)
                    ),
      keras.layers.Conv2D(filters=hp.Int("conv_1_filter",min_value=32,max_value=50,step=16),
                          kernel_size=hp.Choice("conv_1_kernel",values=[3,5,8]),
                          activation="relu"
                    ),
      keras.layers.Flatten(),
      keras.layers.Dense(units=hp.Int("Dense_1_units",min_value=32,max_value=128,step=16),
          activation="relu"),
      keras.layers.Dense(units=10,activation="softmax")


      ])
  model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp.Choice("learning_rate",values=[1e-1,1e-2,1e-3,1e-4])),loss="sparse_categorical_crossentropy",metrics=["accuracy"] )
  return model


```

```python
from keras_tuner import RandomSearch
```

```python
tuner_search=RandomSearch(build_model,objective="val_accuracy",max_trials=5,directory="/content",project_name="Deepak")
```

```python
tuner_search.search(train_iamges,train_labels,epochs=2,validation_split=0.1)
```

```python
best_model=tuner_search.get_best_models(num_models=1)[0]
```

```python
best_model.summary()
```

```python
best_model.fit(train_iamges,train_labels,epochs=5,validation_split=0.1)
```

```python
y_pred=[np.argmax(i) for i in best_model.predict(test_images)]
```
