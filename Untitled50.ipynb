```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df["TotalCharges"]=df["TotalCharges"].apply(lambda x: 0 if x==' ' else x).astype(float)
```

```python
plt.hist(df["tenure"],label=df["Churn"])
```

```python
def print_unique(dataf):
  for columns in dataf:
    if dataf[columns].dtype=="O":
      print(columns,'----',df[columns].unique())
```

```python
print_unique(df)
```

```python
df.replace("No internet service","No",inplace=True)
df.replace("No phone service","No",inplace=True)
```

```python
yes_no_cols=list()
for columns in df:
  if df[columns].dtype=="O":
    for i in (df[columns].unique()):
      if i in "Yes":
        df[columns].replace({"Yes":1,"No":0},inplace=True)
```

```python
print_unique(df[yes_no_cols])
```

```python
df["gender"]=df["gender"].apply(lambda x:1 if x=="male" else 0)
```

```python
df.head()
```

```python
df2=pd.get_dummies(data=df,columns=["InternetService","Contract","PaymentMethod"])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df2.head(3)
```

```python
len(df2.columns)
```

```python
colstotransf=["TotalCharges","MonthlyCharges","tenure"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
df2[colstotransf]=scale.fit_transform(df2[colstotransf])
```

```python
df2.head(4)
```

```python
df2["tenure"].plot()

```

```python
df2["tenure"].describe()
```

```python
df2.head()
```

```python
x=df2.drop(["Churn"],axis="columns")
y=df2["Churn"]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(units=20,input_shape=(26,),kernel_initializer="he_uniform",activation="relu"),
    keras.layers.Dense(units=15,activation="relu"),
    keras.layers.Dense(units=1,activation="sigmoid")

])
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
model.fit(X_train,y_train,epochs=50)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred_final=list()
for i in y_pred:
  if i <0.5:
    y_pred_final.append(0)
  else:
    y_pred_final.append(1)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
tm=confusion_matrix(y_test,y_pred_final)
```

```python
sns.heatmap(tm,annot=True,fmt="d",cmap="viridis")
```

```python
confusion_matrix(y_pred_final,y_test)
```

```python
print(classification_report(y_test,y_pred_final))
```
