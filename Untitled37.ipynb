```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.05,random_state=20)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model=DBSCAN(eps=0.01)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=model.labels_)
```
