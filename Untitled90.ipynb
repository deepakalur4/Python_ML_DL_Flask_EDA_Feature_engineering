```python
import gensim
import pandas as pd
```

```python
df=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Sports_and_Outdoors_5.json.gz",lines=True)
```

```python
df.head()
```

```python
df["reviewText"]
```

```python
reviewText=df["reviewText"].apply(gensim.utils.simple_preprocess)
```

```python
model=gensim.models.Word2Vec(
    window=5,min_count=5,workers=4




)
```

```python
model.build_vocab(reviewText)
```

```python
model.train(reviewText,total_examples=model.corpus_count,epochs=5)
```

```python
model.wv.most_similar("awful")
```

```python
model.wv.similarity('good', 'great')
```

```python
model.wv.similarity('slow','steady')
```
