```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

```python
dir(df)
```

```python
print(df.DESCR)
```

```python
df["data"]
```

```python
df_=pd.DataFrame(df["data"],columns=df["feature_names"])
```

```python
df_.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
scale.fit(df_)
```

```python
sacled_df=scale.fit_transform(df_)
```

```python
sacled_df
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=2)
```

```python
PCA_data=model.fit_transform(sacled_df)
```

```python
model.explained_variance_
```

```python
plt.scatter(PCA_data[:,0],PCA_data[:,1],c=df["target"],cmap="plasma")
```
