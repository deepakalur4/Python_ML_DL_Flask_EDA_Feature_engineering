```python
import seaborn as sn
```

```python
data=sn.load_dataset("tips")
```

```python
data.to_csv("test1.csv")
```

```python
data=session.read.csv("test1.csv",header=True,inferSchema=True)
```

```python
from pyspark.ml.feature import StringIndexer
```

```python
indexer=StringIndexer(inputCol="sex",outputCol="Deepu")
```

```python
data=indexer.fit(data).transform(data)
```

```python
data.printSchema()
```

```python
indexer=StringIndexer(inputCols=["smoker","day","time"],outputCols=["smoker_indexed","day_indexed","time_indexed"])
```

```python
data=indexer.fit(data).transform(data)
```

```python
data.show(5)
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
assemb=VectorAssembler(inputCols=["total_bill","tip","Deepu","smoker_indexed","day_indexed","time_indexed"],outputCol="independed_features")
```

```python
data=assemb.transform(data)
```

```python
data.show(5)
```

```python
data.select("independed_features").show()
```

```python
finalized_data=data.select("independed_features","total_bill")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
train_data,test_data=finalized_data.randomSplit([0.75,0.25])
```

```python
regessor=LinearRegression(featuresCol="independed_features",labelCol="total_bill")
```

```python
regessor=regessor.fit(train_data)
```

```python
regessor.coefficients
```

```python
regessor.intercept
```

```python
pred_res=regessor.evaluate(test_data)
```

```python
pred_res.predictions.show()
```

```python
pred_res.r2
```

```python
pred_res.meanAbsoluteError
```

```python
pred_res.meanSquaredError
```
