```python
class person():
  company="microsoft"
  def __init__(self,a,b):
    self.a=a
    self.b=b

  def info(self):
    print(f"This is the info of {self.a} and {self.b} and he is from {self.company}")

  @classmethod
  def change_company(cls,new_company):
    cls.company=new_company
```

```python
a=person("Deepak",25)
b=person("jyothi",25)
```

```python
a.change_company("india")
```

```python
a.info()
```

```python
b.info()
```

```python
class person():
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def info(self):
    print(f"This is a info of {self.a} and {self.b}")
  @classmethod
  def from_str(cls,string):
    return cls(string.split("-")[0],string.split("-")[1])
```

```python
class com(person):
  def __init__(self,c,d,a,b):
    super().__init__(a,b)
    self.c=c
    self.d=d
  def func(self):
    super().info()
```

```python
a=com("Deepak",25,"jyo",35)
```

```python
a.func()
```
