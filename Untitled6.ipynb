```python
def leap(year):
  if year%4==0:
    if year%100==0:
      if year%400==0:
        print("Its a leap year")
      else:
        print("Its a not leap year")
    else:
      print("Its a leap year")
  else:
    print("Its a not leap year")
```

```python
leap(1900)
```

```python
def leap(year):
  check={True:"leap year",False:"Not leap"}
  return check[year%4==0 and (year%100!=0 or year%4000==0)]
```

```python
leap(1990)
```

```python
def print_formatted(number):

    width = len("{0:b}".format(number))

    for i in range(1, number + 1):
        print("{0:{w}d} {0:{w}o} {0:{w}X} {0:{w}b}".format(i, w = width))
```

```python
print_formatted(5)
```
