```python
import spacy,pandas as pd
```

```python
spacy.cli.download("en_core_web_lg")
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/OmkarPathak/Playing-with-datasets/master/Email%20Spam%20Filtering/emails.csv")
```

```python
df=df.sample(200,random_state=20).reset_index()
```

```python
df.head()
```

```python
df["text"]=df["text"].apply(lambda x:' '.join(x.split()[1:]))
```

```python
df.head()
```

```python
import tensorflow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
df["text"]=df["text"].apply(lambda x :[one_hot(x,500)])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
df["text"]=df["text"].apply(lambda x :pad_sequences(x,maxlen=500,padding="post"))
```

```python
df.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
model=Sequential([
    Embedding(500,2,input_length=500),



])
```

```python
model.compile(optimizer="Adam",loss="binary_cross_entropy")
```

```python
model.summary()
```

```python
model.predict(df["text"][0]).reshape(500,2)
```
