```python
import pandas_datareader as pdr
import pandas as pd
import os
from datetime import datetime
```

```python
import yfinance as yf
```

```python
yf.pdr_override()
```

```python
df_tesla=pdr.data.get_data_yahoo("TSLA")
```

```python
df_tesla.head()
```

```python
df_tesla["Open"].plot(figsize=(15,6))
```

```python
df_tesla["Open: 10 Days rolling"]=df_tesla["Open"].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla.head()
```

```python
df_tesla[["Open","Open: 10 Days rolling"]].plot(figsize=(15,6))
```

```python
df_tesla[["Open","Open: 10 Days rolling"]].plot(xlim=["2021-01-01","2024-01-01"],figsize=(15,6))
```

```python
df_tesla["Open: 30 Days rolling"]=df_tesla["Open"].rolling(window=30,min_periods=1).mean()
df_tesla["Open: 40 Days rolling"]=df_tesla["Open"].rolling(window=40,min_periods=1).mean()
df_tesla["Open: 50 Days rolling"]=df_tesla["Open"].rolling(window=50,min_periods=1).mean()
df_tesla["Open: 60 Days rolling"]=df_tesla["Open"].rolling(window=60,min_periods=1).mean()
```

```python
df_tesla[["Open","Open: 30 Days rolling","Open: 40 Days rolling","Open: 50 Days rolling","Open: 60 Days rolling"]].plot(xlim=["2021-01-01","2024-01-01"],figsize=(15,6))
```

```python
df_tesla["Open"].expanding().mean()
```

```python
df_tesla["Open"].expanding().mean().plot(figsize=(15,6))
```

```python
df_tesla["Open: Exponential weighted moving avg"]=df_tesla["Open"].ewm(alpha=0.1).mean()
```

```python
df_tesla[["Open","Open: Exponential weighted moving avg"]].plot(xlim=["2021-01-01","2024-01-01"],figsize=(15,6))
```
