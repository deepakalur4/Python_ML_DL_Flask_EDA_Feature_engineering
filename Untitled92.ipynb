```python
import tensorflow as tf
```

```python
data=[1,2,3,-5,6,-10,11,13,25]
```

```python
dataset=tf.data.Dataset.from_tensor_slices(data).filter(lambda x:x>0).map(lambda x:x*10).shuffle(2).batch(2)
```

```python
for i in dataset.as_numpy_iterator():
  print(i)
```
