```python
class shape():
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def area(self):
    return self.x*self.y

class circle(shape):
  def __init__(self,radius):
    self.radius=radius
    super().__init__(radius,radius)

  def area(self):
    return 3.14*super().area()
```

```python
a=circle(5)
```

```python
a.area()
```

```python
class vec1():
  def __init__(self,x,y,z):
    self.x=x
    self.y=y
    self.z=z
  def info(self):
    print("this is the vector (self.x)")

  def __repr__(self):
    return (f"This is the vector of {self.x}x+{self.y}y+{self.z}z")

  def __add__(self,p):
    return f"{self.x+p.x}x+{self.y+p.y}y+{self.z+p.z}z"
```

```python
a=vec1(1,2,3)
```

```python
b=vec1(5,6,7)
```

```python
print(a+b)
```
