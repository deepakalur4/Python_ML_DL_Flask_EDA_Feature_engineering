#K means clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
X, y = make_blobs(n_samples=1000, centers=3, n_features=2,random_state=23)
```

```python
X=np.random.rand(500,2)
y=np.random.rand(500,2)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++')
  kmeans.fit(X_train)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
```

```python
from kneed import KneeLocator
```

```python
kl = KneeLocator(range(1,11), wcss, curve="convex", direction="decreasing")
```

```python
kl.elbow
```

```python
kl.plot_knee()
```

```python
  kmeans=KMeans(n_clusters=4,init='k-means++')
  y_labbles=kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labbles)
```

```python
y_test_labbles=kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test_labbles)
```

```python
from sklearn.metrics import silhouette_score
```

```python
coeff=[]
for i in range(2,11):
  kmeans=KMeans(n_clusters=i,init='k-means++')
  kmeans.fit(X_train)
  score=silhouette_score(X_train,kmeans.fit(X_train).labels_)
  coeff.append(score)

```

```python
plt.plot(range(2,11),coeff)
plt.xticks(range(2,11))
```

#Heirarchal clustering

```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=datasets.load_iris()
```

```python
df_data=pd.DataFrame(df["data"])
```

```python
df=datasets.load_iris()
df_data.columns=df.feature_names
```

```python
df_data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
scaled=scale.fit_transform(df_data)
```

```python
scaled.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled=pca.fit_transform(scaled)
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
```

```python
cluster.fit(pca_scaled)
```

```python
y=cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=y)
```

```python
from scipy.cluster import hierarchy as sc
```

```python
Z = hierarchy.linkage(pca_scaled, method='ward')
plt.figure(figsize=(20,7))
plt.title("Dendrograms")
dendrogram = hierarchy.dendrogram(Z)
```

```python
coeff=[]
for i in range(2,11):
  cluster=AgglomerativeClustering(n_clusters=i, metric='euclidean', linkage='ward')
  a=cluster.fit(pca_scaled)
  score=silhouette_score(pca_scaled,a.labels_)
  coeff.append(score)
```

```python
coeff
```

```python
plt.plot(range(2,11),coeff)
plt.xticks(range(2,11))
```
