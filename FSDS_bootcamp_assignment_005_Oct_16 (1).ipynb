#### 1. On amazon website we have the following available items - 

```python
amazon_website = [("watch", 5000), ("phone", 10000), ("laptop", 50000), ("shirt", 1000)]
```
Take the user input N no. of times and find out the total cost to be paid?

OR

You can ask the user quntity or the count of item?

if the item is not present in the amazon website then print that item as out of stock and do add it total


HINT: while, for loop.. indexing

````python
amazon_website = [("watch", 5000), ("phone", 10000), ("laptop", 50000), ("shirt", 1000)]
total_amount=0
while True:
  item=tuple(input("Please enter the item and the price you want to add to the list or q to finalise the list : \n").split(" "))
  if item[0] != "q":
    if item[0] not in amazon_website[0]:
       amazon_website.append(item)
       print(f"The new list is: {amazon_website}")
    else:
      print(f"The {item[0]} is already present in the list: {amazon_website}")
  else:
    for i in range(len(amazon_website)):
      total_amount=total_amount+int(amazon_website[i][1])
    print(f"Thanks for the shopping, The total payable amount is : {total_amount}")
    break
````

#### 3. Write a Python program to find the repeated items of a tuple.

````python
values=(1,22,33,44,66,88,88,88,44,1,22)
repeated_items = set()
for i in values:
  if values.count(i)>1:
    repeated_items.add(i)
print(f"The repeated items in the tuple are {tuple(repeated_items)}")
````
