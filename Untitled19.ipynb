```python
import pandas as pd
```

```python
df=[1,1,10,4,19,177,200]
df=sorted(df)
```

```python
outlires=list()
```

```python
def detect_out(func):
  threshould=2
  mean=np.mean(df)
  std_dev=np.std(df)

  for i in func:
    # print(i)
    z=((i-mean))/std_dev
    print(np.abs(z))
    if np.abs(z)>threshould:
      print(np.abs(z),threshould)
      outlires.append(i)
  return outlires
```

```python
detect_out(df)
```
