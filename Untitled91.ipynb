```python
import tensorflow as tf
```

```python
daily_summary_no=[1,2,3,100,400,-5,3,7,6]
```

```python
df=tf.data.Dataset.from_tensor_slices(daily_summary_no)
df=df.filter(lambda x:x>0)
```

```python
for i in df:
  print(i)
```

```python
for i in df.batch(3):
  print(i)
```

```python
daily_summary_no=[1,2,3,100,400,-5,3,7,6]
```

```python
tf_data=tf.data.Dataset.from_tensor_slices(daily_summary_no)
```

```python
tf_data=tf_data.filter(lambda x:x>0).map(lambda x:x*10).shuffle(2).batch(2)
```

```python
for i in tf_data:
  print(i)
```
