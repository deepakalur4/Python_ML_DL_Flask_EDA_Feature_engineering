```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(r"https://raw.githubusercontent.com/krishnaik06/Complete-Deep-Learning/master/ANN/Churn_Modelling.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,3:-1]
y=df.iloc[:,-1]
```

```python
Gender=pd.get_dummies(df["Gender"])
Geography=pd.get_dummies(df["Geography"])
```

```python
x=pd.concat([x,Gender,Geography],axis=1)
```

```python
x.head()
```

```python
x.drop(["Geography","Gender"],axis=1,inplace=True)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
classifer=Sequential()
```

```python
X_train.shape
```

```python
classifer.add(Dense(units=6,kernel_initializer="he_uniform",activation="relu",input_dim=13))
```

```python
classifer.add(Dense(units=6,kernel_initializer="he_uniform",activation="relu"))
classifer.add(Dropout(0.3))
```

```python
classifer.add(Dense(units=1,kernel_initializer="glorot_uniform",activation="sigmoid"))
```

```python
classifer.summary()
```

```python
classifer.compile(optimizer="adam",loss="BinaryCrossentropy",metrics=["accuracy"])
```

```python
classifer.summary()
```

```python
model=classifer.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
y_pred=classifer.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```
