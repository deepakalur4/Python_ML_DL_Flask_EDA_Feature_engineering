#Under_Sampling of majority clas

```python
class_0,class_1=df["Churn"].value_counts()
```

```python
df["Churn"]=df["Churn"].apply(lambda x:1 if x=="Yes" else 0)
```

```python
df1=df[df["Churn"]==0].sample(1869,random_state=100)
df2=df[df["Churn"]==1].sample(1869,random_state=100)
```

```python
df=pd.concat([df1,df2],axis=0)
```
