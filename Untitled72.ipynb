```python
import spacy,pandas as pd
```

```python
spacy.cli.download("en_core_web_lg")
```

```python
nlp=spacy.load("en_core_web_lg")
```

```python
df=pd.read_json("https://raw.githubusercontent.com/codebasics/nlp-tutorials/main/14_word_vectors_spacy_text_classification/news_dataset.json")
```

```python
df.head()
```

```python
df.shape
```

```python
df["category"].value_counts()
```

```python
df["category"]=df["category"].apply(lambda x:0 if x=="CRIME" else (1 if x=="SPORTS" else 3) )
```

```python
df["text"]=df["text"].apply(lambda x:nlp(x).vector)
```

```python
df["text"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,_train, y_test = train_test_split(df[["text"]].values,df["category"],
                                   random_state=104,
                                   test_size=0.2,
                                   shuffle=True)
```

```python
X_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(X_train,_train)
```
