```python
import tensorflow_hub as hub
import tensorflow_text as text
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/master/47_BERT_text_classification/spam.csv")
```

```python
df.head()
```

```python
df["Category"].value_counts()
```

```python
ham=df[df["Category"]=="ham"]
spam=df[df["Category"]=="spam"]
```

```python
ham=ham.sample(747,random_state=20)
```

```python
df=pd.concat([ham,spam],axis=0)
```

```python
df["Category"].value_counts()
```

```python
df.head()
```

```python
df["Category"]=df["Category"].apply(lambda x:1 if x=="spam" else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,ytest=train_test_split(df["Message"],df["Category"],test_size=0.2,random_state=42,stratify=df["Category"])
```

```python
X_train[:3]
```

```python
preprocessor = hub.KerasLayer("https://kaggle.com/models/tensorflow/bert/frameworks/TensorFlow2/variations/en-uncased-preprocess/versions/3")
encoder = hub.KerasLayer("https://www.kaggle.com/models/tensorflow/bert/frameworks/TensorFlow2/variations/bert-en-uncased-l-10-h-128-a-2/versions/2")
```

```python
def get_sentence_emb(sentence):
  sent_preprocess=preprocessor(sentence)
  word_emb=encoder(sent_preprocess)
  return word_emb["pooled_output"]
```

```python
a=get_sentence_emb(["he is a good boy","she is a good girl"])
```

```python
a
```

```python
import keras,tensorflow as tf
```

```python
inpout_layer=keras.layers.Input(shape=(),dtype=tf.string,name="text")
pre=preprocessor(inpout_layer)
output=encoder(pre)
```

```python
nueral=keras.layers.Dropout(0.1,name="Dropout")(output["pooled_output"])
nueral_2nd_layer=keras.layers.Dense(1,activation="sigmoid",name="desne_alyer")(nueral)
```

```python
model=keras.Model(inputs=[inpout_layer],outputs=[nueral_2nd_layer])
```

```python
model.summary()
```

```python
model.compile("Adam","binary_crossentropy",metrics=[keras.metrics.BinaryAccuracy(name='binary_accuracy')])
```

```python
model.fit(X_train,y_train,epochs=5)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=[0 if i<0.5 else 1 for i in y_pred.reshape(-1,)]
```

```python
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
sns.heatmap(confusion_matrix(ytest,y_pred),fmt="d",annot=True,cmap="viridis")
plt.xlabel("predicted")
plt.ylabel("truth")
```
