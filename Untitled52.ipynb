```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import tensorflow as td
from  tensorflow.keras import models,layers,datasets
```

```python
(X_train,y_train),(X_test,y_test)=datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
#ANN
model=models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(units=40,kernel_initializer="he_uniform",activation="relu"),
    layers.Dense(units=30,kernel_initializer="he_uniform",activation="relu"),
    layers.Dense(units=10,kernel_initializer="glorot_uniform",activation="softmax"),


])
model.compile(optimizer="Adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=[np.argmax(i) for i in model.predict(X_test)]
```
