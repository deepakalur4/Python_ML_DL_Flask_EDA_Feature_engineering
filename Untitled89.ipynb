```python
import gensim
```

```python
import pandas as pd
```

```python
df=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Cell_Phones_and_Accessories_5.json.gz", lines=True)
```

```python
df.shape
```

```python
x=df["reviewText"]
```

```python
x[0]
```

```python
review_text=df["reviewText"].apply(gensim.utils.simple_preprocess)
```

```python
review_text
```

```python
moel=gensim.models.Word2Vec(
    window=10,min_count=4,workers=4




)
```

```python
moel.build_vocab(review_text)
```

```python
moel.train(review_text,total_examples=moel.corpus_count,epochs=moel.epochs)
```

```python
moel.save("Deepak.model")
```

```python
moel.wv.most_similar("bad")
```

```python
moel.wv.similarity(w1="good",w2="great")
```
